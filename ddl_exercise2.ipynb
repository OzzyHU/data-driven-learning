```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('class_data.csv')
df.head()
```

```python
df_subset = df[df['instagram_followers'] < 500] # subsetting dataframe to make it more specific and visually appealing
df_subset

df_subset['instagram_followers'].name = 'Followers'  #Labeling the variable for plotting
ig_followers = df_subset['instagram_followers'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(ig_followers, kde = False) # Seaborn displot function
plt.title('Instagram followers')
plt.show()
```

```python
mean_size = df['instagram_followers'].mean()
median_size = df['instagram_followers'].median()

print(f'The median of IG followers is: {median_size}')
print(f'The mean of IG followers is: {mean_size}')
```

```python
sns.distplot(ig_followers, kde = False) #I need to drop the NA's, otherwise Seaborn won't plot
plt.title('Instagram followers')
plt.axvline(median_size, 0, 100, color='green', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_size, 0, 100, color='blue', label='mean') 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

```python
df['linkedin_con'].name = 'Connections'  #Labeling the variable for plotting
linkedin_connections = df['linkedin_con'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(linkedin_connections) # Seaborn displot function
plt.title('LinkedIn Connections')
plt.ylabel('Proportion')
plt.show()
```

```python
phones_owned = df['phones_owned'].dropna()
sns.distplot(phones_owned, rug=True)
plt.title('Phones owned by the users')
plt.xlabel('Amount of phones)')
plt.ylabel('Proportion')
plt.show()
```

```python
texts_sent = df['texts_yesterday'].dropna()
sns.boxplot(texts_sent)
plt.title('Texts send yesterday')
plt.xlabel('Texts sent')
plt.show()
```

```python
phone_apps = df[df['apps_phone'] < 1000]#Subsetting the dataframe to include only more than 5 apps
sns.violinplot(phone_apps['apps_phone'])
sns.swarmplot(phone_apps['apps_phone'], color='orange') #default color doesn't stand out
plt.title('Amount of apps on phones of students')
plt.ylabel('Students')
plt.xlabel('Apps')
plt.show()
```
