```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.scatterplot(x='PetalLengthCm',y='SepalLengthCm',data=iris, style= 'Species')
plt.xlabel('Petal Length cm')
plt.ylabel('Sepal Length cm')
plt.show()
```

```python
km = KMeans(n_clusters = 3) #create a new k-means model with 3 clusters
X = iris[['SepalLengthCm', 'PetalLengthCm']] #get the X variables from the dataframe
km = km.fit(X) #calculate the cluster centers
iris['cluster'] = km.predict(X) #predict the clusters of each observation and store in the dataframe
iris.head()
```

```python
sns.scatterplot(x='PetalLengthCm',y='SepalLengthCm',data=iris, style='cluster', hue='Species')
plt.xlabel('Petal Length cm')
plt.ylabel('Sepal Length cm')
plt.show()
```

The dataset works the best with 2 clusters, since this is the most obvious clustering and based on the data it's the best option.
