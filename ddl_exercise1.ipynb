```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('survey_data.csv')
df1.head()
```

```python
# Separating the data with semicoloms in order to make it compatible with the other csv file
df2 = pd.read_csv('steps_data.csv', sep = ';')
df2.head()
```

```python
# Looking for commong identifier to merge both files into a new dataframe
df3 = pd.merge(df1, df2, on = 'id')
df3.head()
```
