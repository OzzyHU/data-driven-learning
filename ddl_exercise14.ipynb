```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
df3 = pd.merge(movies, ratings, on = 'movieId')
df3.head()
```

```python
movie_ratings = pd.pivot_table(data=df3, index='userId', columns='title', values='rating')
movie_ratings.head()
```

```python
new_movie_ratings = movie_ratings.reindex(movie_ratings.count().sort_values(ascending=False).index, axis=1).fillna(0)
new_movie_ratings.head()
```

```python
km = KMeans(n_clusters = 4) 
X = new_movie_ratings
km = km.fit(X) 
movie_ratings['cluster'] = km.predict(X) 
movie_ratings.head()
```

```python
movie_ratings['cluster'].value_counts()
```

```python
movie_ratings = movie_ratings.reindex(movie_ratings.count().sort_values(ascending=False).index, axis=1)
movie_ratings.head()
```

```python
cluster_mean = movie_ratings.groupby(movie_ratings['cluster']).mean()
cluster_mean 
```

```python
cluster_0 = cluster_mean.iloc[0]
cluster_0 = cluster_0.sort_values(ascending=False)
cluster_0.head(4)
```
